```python
import pandas as pd
```

```python
vendas_df = pd.read_csv('/content/Vendas - Intro Pandas.csv', sep = ";")
```

```python
vendas_df
```

```python
vendas_df.head()
```

```python
vendas_df.tail()
```

```python
vendas_df = vendas_df[['Cod Item','Data Faturamento', 'Quantidade', 'Cod Cliente']]
```

```python
vendas_df
```

```python
freq_cliente = vendas_df['Cod Cliente'].value_counts()
```

```python
freq_cliente[:5]
```

```python
freq_produto = vendas_df['Cod Item'].value_counts()
```

```python
freq_produto[:5]
```

```python
freq_produto[:5].plot.bar(x='Cod Item', y='Quantidade')
```

```python
produto_df = vendas_df[['Cod Item', 'Quantidade']].groupby('Cod Item').sum()
```

```python
produto_df
```

```python
produto_df = produto_df.sort_values('Quantidade', ascending=False)
```

```python
produto_df
```

```python
qtd_mais_vendidos = produto_df['Quantidade'].max()
prod_mais_vendido = produto_df['Quantidade'].idxmax()

print(f'O produto mais vendido foi {prod_mais_vendido}, com {qtd_mais_vendidos} unidades')
```

```python
qtd_menos_vendidos = produto_df['Quantidade'].min()
prod_menos_vendido = produto_df['Quantidade'].idxmin()

print(f'O produto menos vendido foi {prod_menos_vendido}, com {qtd_menos_vendidos} unidades')
```
